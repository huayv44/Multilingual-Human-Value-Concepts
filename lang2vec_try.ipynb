```python
import lang2vec.lang2vec as l2v
```

```python
import torch
```

```python
lang_json = torch.load("lang.pt")
```

```python
lang_code_3 = [lang_json[code_2]["code_3"] for code_2 in lang_json]
```

```python
l2v.distance("featural", *lang_code_3)
```

```python
type_str = """genetic
geographic
syntactic
inventory
phonological
featural"""
```

```python
type_lst = type_str.split()
```

```python
type_lst
```

```python
res = l2v.distance(type_lst, *lang_code_3)
```

```python
torch.save(res, "lang_sim.pt")
```

```python
res
```

```python
lang_score = {}
for l1_idx, lang1 in enumerate(lang_json):
    lang_score[lang1] = {}
    for l2_idx, lang2 in enumerate(lang_json):
        lang_score[lang1][lang2] = {}
        for t_idx,t in enumerate(type_lst):
            lang_score[lang1][lang2][t] = res[t_idx][l1_idx][l2_idx].item()
```

```python
torch.save(lang_score, "lang_sim.pt")
```
